```python
# The followings are codes for STA2004 HW2

import numpy as np
import matplotlib.pyplot as plt

#The followings are codes for Q1, the output is the figure of L(lambda)-L(lambda_hat)
for num in (1,5,10,30):                              # n can be selected among[1,5,10,30]
    y = np.random.exponential(10, num)    # y follows exponential distribution, with mean=10, amount=num
    lambda_list = np.arange(0.01, 1, 0.01)  # store lambda, ranges from 0.01 to 1, into a list
    lambda_hat = 1/10                       # lambda_hat = 1/y_bar
    y_bar = 10
    difference = []
    for element in lambda_list:
        L_lambda = np.power(element, num) * np.exp(-1 * element * num * y_bar)                    # L(lambda) is the cumulative product of (lambda*exp(-1*lambda*y))
        L_lambda_hat = np.power(lambda_hat, num) * np.exp(-1 * lambda_hat * num * y_bar)
        difference.append(L_lambda - L_lambda_hat)
    plt.plot(lambda_list, difference)
    plt.title('n={}'.format(num))
    plt.show()
```

```python
lambda_bar = 0.1    # lambda_bar = 1/y_bar
Z_half_alpha = 1.96
for n in (1, 5, 10, 30):
    root_n = np.sqrt(n)
    half_length = Z_half_alpha * lambda_bar / root_n
    print('array([', lambda_bar-half_length, ',', lambda_bar+half_length, ']),')
```

```python
#The followings are codes for Q3
lambda_zero = 1
Z_half_alpha = 1.96
size = 10000
plt.title('Confidence Intervals')
plt.xlabel('lambda')
plt.ylabel('Samples')
plt.vlines(1, 0, 100, colors='green')
for num in range(100):
    data_set = np.random.exponential(lambda_zero, size)
    mean_of_data = np.mean(data_set)
    lambda_hat = 1/mean_of_data
    half_length = lambda_hat * Z_half_alpha/np.sqrt(size)
    if lambda_hat - half_length > 1 or lambda_hat + half_length < 1:
        plt.hlines(num, lambda_hat - half_length, lambda_hat + half_length, colors='red')
    else:
        plt.hlines(num, lambda_hat - half_length, lambda_hat + half_length, colors='black')
plt.show()
```
